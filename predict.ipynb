```python
def convertrange(x):
    temp=x.split('-')
    if(len(temp)==2):
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data['total_sqft']=data['total_sqft'].apply(convertrange)
```

```python
def removefalsesqft(df):
    df_output=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.pricepersft)
        
        st=np.std(subdf.pricepersft)
        
        gen_df=subdf[(subdf.pricepersft > (m-st)) & (subdf.pricepersft <= (m+st))]
        df_output=pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
data=removefalsesqft(data)
data.describe()
```

```python
data.head()
```

```python
def bhkfalseremove(df):
    exindices=np.array([])
    for loc,locdf in df.groupby('location'):
        bhkstats={}
        for bhk,bdf in locdf.groupby('bhk'):
            bhkstats[bhk]={
                'mean':np.mean(bdf.pricepersft),
                'std':np.std(bdf.pricepersft),
                'count':bdf.shape[0]
            }
#        print(loc,bhkstats)
        for bhk,bdf in locdf.groupby('bhk'):
            stats=bhkstats.get(bhk-1)
            if stats and stats['count']>5:
                exindices=np.append(exindices,bdf[bdf.pricepersft<(stats['mean'])].index.values)
    return  df.drop(exindices,axis='index')
```

```python
data=bhkfalseremove(data)
```

```python
X=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Applying Linear Regression

```python
column_trans=make_column_transformer((OneHotEncoder(sparse=False),['location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression(normalize=True)
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

## applying Lasso

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso=pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

## applying Ridge

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge=pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization - ",r2_score(y_test,y_pred_lr))
print("lasso - ",r2_score(y_test,y_pred_lasso))
print("ridge - ",r2_score(y_test,y_pred_ridge))
```
